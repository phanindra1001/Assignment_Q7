```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("P:/python/Q7.csv")
```

```python
data
```

```python
print('MEAN:\n',data.mean())
print('MEDIAN:\n',data.median())
print('STANDARD DEVIATION:\n',data.std())
print(f'MODE:\n Points:\n {data.Points.mode()}\nScore:\n{data.Score.mode()}\nWeight:\n{data.Weigh.mode()} ')
print('VARIANCE:\n',data.var())
points = data.Points.max()-data.Points.min()
print('RANGE')
print(f'range for points: {points}')
score = data.Score.max()-data.Score.min()
print(f'range for score: {score}')
weight = data.Weigh.max()-data.Weigh.min()
print(f'range for weight: {weight}')
```

```python
plt.subplots(figsize=(25,10))
plt.subplot(1,3,1)
plt.boxplot(data.Points, vert = False)
plt.title('Points')
plt.subplot(1,3,2)
plt.boxplot(data.Score, vert = False)
plt.title('Score')
plt.subplot(1,3,3)
plt.boxplot(data.Weigh, vert = False)
plt.title('Weight')
plt.show()
```
